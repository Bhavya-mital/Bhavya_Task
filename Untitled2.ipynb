columns such as mainroom,guestroom,basement,hotwaterheating,airconditioning,parking contains some data in wrong format that was accordingly replaced with the correct format data as needed by the model

```python
wtrue=["Of course" ,"Indeed", "yep" ,"Haan" ,"True" ,"y", "Y", "['y']","Yo","Haanji"]
rtrue='yes'
wfalse=["Nope","Nakko","False" ,"Naaji", "N" ,"n","ille","N","['n']","Nah"]
rfalse='no'
df['mainroad']=df['mainroad'].replace(wtrue,rtrue)
df['mainroad']=df['mainroad'].replace(wfalse,rfalse)
df['guestroom']=df['guestroom'].replace(wtrue,rtrue)
df['guestroom']=df['guestroom'].replace(wfalse,rfalse)
df['basement']=df['basement'].replace(wtrue,rtrue)
df['basement']=df['basement'].replace(wfalse,rfalse)
df['hotwaterheating']=df['hotwaterheating'].replace(wtrue,rtrue)
df['hotwaterheating']=df['hotwaterheating'].replace(wfalse,rfalse)
df['airconditioning']=df['airconditioning'].replace(wtrue,rtrue)
df['airconditioning']=df['airconditioning'].replace(wfalse,rfalse)
df['parking']=df['parking'].replace(wtrue,rtrue)
df['parking']=df['parking'].replace(wfalse,rfalse)
df['prefarea']=df['prefarea'].replace(wtrue,rtrue)
df['prefarea']=df['prefarea'].replace(wfalse,rfalse)
df
```

for numeric column best approach is to extract the most common value along with its frequency to understand the variation.for this below function is made

```python
for col in ["area","bedrooms","bathrooms","stories","parking"]:
    print(col)
    print(df[col].value_counts(normalize=True).sort_values(ascending=False).head())
    print("---------")
```

interpolate method is used for the numeric columns to fill null values with an estimated value based on the immediate value.along with this i used fillna method to replace the missing blocks left after interpolation with the most ocurred value

```python
df['area']=df['area'].interpolate(method='linear').fillna(1).astype(int)
df['bedrooms']=df['bedrooms'].interpolate(method='linear').fillna(3).astype(int)
df['bathrooms']=df['bathrooms'].interpolate(method='linear').fillna(1).astype(int)
df['stories']=df['stories'].interpolate(method='linear').fillna(2).astype(int)
df['parking']=df['parking'].interpolate(method='linear').fillna(0).astype(int)
df
```

among the rest column mainroad and hotwaterheating have pattern that was most repeated with yes or no.so accordingly the null values have been placed.

```python
df['mainroad'].fillna('yes',inplace=True)
df['hotwaterheating'].fillna('no',inplace=True)
df
```

for the rest columns no such above pattern was observed so i changed them to numeric value using replace method and then observed the frequency of the most occured

```python
df['guestroom'].replace(['yes','YES','no','NO'],[1,1,0,0],inplace=True)
df['basement'].replace(['yes','YES','no','NO'],[1,1,0,0],inplace=True)
df['airconditioning'].replace(['yes','YES','no','NO'],[1,1,0,0],inplace=True)
df['prefarea'].replace(['yes','YES','no','NO'],[1,1,0,0],inplace=True)
df['furnishingstatus'].replace(['furnished','semi-furnished','unfurnished'],[1,0.5,0],inplace=True)
df
```

```python
for col in ["guestroom","airconditioning","basement","prefarea","furnishingstatus"]:
    print(col)
    print(df[col].value_counts(normalize=True).sort_values(ascending=False).head())
    print("---------")
```

replacing null with most occured gives a chance of 99% correctness and with any number gives a chance of 0% correctness

```python
df["guestroom"].fillna(0, inplace = True)
df["airconditioning"].fillna(0, inplace = True)
df["basement"].fillna(0, inplace = True)
df["prefarea"].fillna(0, inplace = True)
df["furnishingstatus"].fillna(0.5, inplace = True)
df
```
