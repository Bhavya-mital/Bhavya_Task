the below histograms reveal the variation of individual column values to the number of houses for a more clear picture of distribution.

```python

plt.hist(df['X2 house age'],bins=10)  # Adjust the number of bins as needed
plt.xlabel('house age')  # Replace with a label for the x-axis
plt.ylabel('number')  # Replace with a label for the y-axis
plt.show()
plt.hist(df['X3 distance to the nearest MRT station'],bins=10)  #
plt.xlabel('distance from nearest metro')  
plt.ylabel('number')  
plt.show()
plt.hist(df['X4 number of convenience stores'],bins=10)  
plt.xlabel('stores nearby')  
plt.ylabel('number')  
plt.show()
plt.hist(df['X5 latitude'],bins=10)  
plt.xlabel('latitude')  
plt.ylabel('number')  
plt.show()
plt.hist(df['X6 longitude'],bins=10)  
plt.xlabel('longitude')  
plt.ylabel('number')  
plt.show()
plt.hist(df['Y house price of unit area'],bins=10)  
plt.xlabel('price of unit area') 
plt.ylabel('number')  
plt.show()
```

the below given bar chart depicts that all the houses provided have almost same variation of longitude with latitude within all the values.thus almost same area can be assumed

```python


column1 = 'X6 longitude'
column2 = 'X5 latitude'
plt.bar(df[column1], df[column2])
plt.xlabel(column1)
plt.ylabel(column2)
plt.title('Vertical Bar Chart Between {} and {}'.format(column1, column2))

# Rotate x-axis labels for better readability if necessary
plt.xticks(rotation=45)
plt.show()
```
