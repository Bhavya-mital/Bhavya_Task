the dropna() method in pandas returns a new DataFrame and will not change the original.the below given output represents the rows that does not contain empty cells in the original data set.

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\DELL\Desktop\Task_4.csv")

new_df = df.dropna()

print(new_df.to_string())
```

If want to change the original DataFrame use the inplace = True argument. dropna(inplace = True) in pandas will NOT return a new DataFrame but  will remove all rows containing NULL values from the original .

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\DELL\Desktop\Task_4.csv")

df.dropna(inplace = True)

print(df.to_string())
```

.fillna method in pandas replace null values with the values 150.instead to deleting rows with null value we can replace it with certain values.

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\DELL\Desktop\Task_4.csv")

df.fillna(150, inplace = True)
print(df.to_string())
```

```python
import pandas as pd
df = pd.read_csv(r"C:\Users\DELL\Desktop\Task_4.csv")

# Define a custom function to clean the data
def clean_data(value):
    if pd.isna(value):
        return 'No'  # Fill empty cells with 'No'
    elif 'yes' in str(value).lower():
        return 'Yes'
    elif 'no' in str(value).lower():
        return 'No'
    else:
        return 'No'  # Replace any other words with 'No'

# Apply the custom function to clean the specific column
df['hotwaterheating'] = df['hotwaterheating'].apply(clean_data)

# Save the cleaned DataFrame to a new or the same CSV file
df.to_csv(r"C:\Users\DELL\Desktop\Task_4.csv", index=False)
print(df.to_string())



```

To discover duplicates use the duplicated() method.it will return true if any row discovered will be a duplicate row otherwise will return false as shown.dtype.bool at end of output tells that the data type given boolean values that  are true or false

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\DELL\Desktop\Task_4.csv")

print(df.duplicated())
```

the .drop_duplicates method removes all the duplicated rows from the dataset. inplace = True method does NOT return a new Data but  will remove all duplicates from the original Data. as our given data does not have any duplicate so none of the row is removed .

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\DELL\Desktop\Task_4.csv")

df.drop_duplicates(inplace = True)

print(df.to_string())
```
